載入套件

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


from sklearn import decomposition
from sklearn import datasets

np.random.seed(5)
%matplotlib inline
```

載入 iris 資料集

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

設定 模型 估計參數[參數參考](https://blog.csdn.net/u012162613/article/details/42192293)

```python
centers = [[1, 1], [-1, -1], [1, -1]]
pca = decomposition.PCA(n_components=3)
```

資料建模 並 視覺化 結果

```python
pca.fit(X)
X = pca.transform(X) #降維後的數據, 三個特徵
```

[Axes3D](https://matplotlib.org/3.1.0/api/_as_gen/mpl_toolkits.mplot3d.axes3d.Axes3D.html)

```python
# 做一個空的圖
fig = plt.figure(1, figsize=(4, 3))
plt.clf() # clears the entire current figure
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()
```

```python
# 做一個空的圖
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()

for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))  
    # 標示不同群的名字
    #'y==0'得到的，是一個一維的true/false列表
    #所以X[y==0,0]就是X[y==0]第一個column的值，也就是類別為0的第一個特徵的值
    #取平均，則可以得到類別為0的資料點中，第一項特徵的平均值
    #我們要將"setosa"這個文字串顯示在類別0的資料集旁，就是要讓它顯示在這群資料中心點的附近，ax.text3D(第一維度,第二維度,第三維度,字串內容)
    
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
