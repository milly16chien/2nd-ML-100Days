```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Merge / Transform

```python
# 生成範例用的資料 ()
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])

df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
print(df1)
print(df2)
print(df3)
print(df4)
```

```python
# 沿縱軸合併 (row bind)
result = pd.concat([df1, df2, df3])
result
```

```python
# 沿橫軸合併  column bind
result = pd.concat([df1, df4], axis = 1)
result
```

```python
# 沿橫軸合併 沒有na的欄位就會串起來
result = pd.concat([df1, df4], axis = 1, join = 'inner') # 硬串接
print(result)
```

```python
# 沿橫軸合併  只顯示沒有重複的部分
result = pd.merge(df1, df4, how='inner')
print(result)
```

```python
# 將 欄-列 逐一解開
print(df1)
df1.melt()
```

## Subset

```python
# 設定 data_path
dir_data = './data/'
f_app = os.path.join(dir_data, 'application_train.csv')
#print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 取 TARGET 為 1 的
sub_df = app_train[app_train['TARGET'] == 1] # target = 1 的 row 為true, 其餘為false
sub_df.head()
```

```python
# 取 SK_ID_CURR, TARGET 兩欄 ，AMT_INCOME_TOTAL 大於平均值的資料
sub_df = app_train.loc[app_train['AMT_INCOME_TOTAL'] > app_train['AMT_INCOME_TOTAL'].mean(), ['SK_ID_CURR', 'TARGET']]
sub_df.head()
```

## Groupby

```python
# 計算各組數量['AMT_INCOME_TOTAL']
app_train.groupby(['NAME_CONTRACT_TYPE']).size()
```

```python
# 計算 NAME_CONTRACT_TYPE各組的AMT_INCOME_TOTAL統計量
app_train.groupby(['NAME_CONTRACT_TYPE'])['AMT_INCOME_TOTAL'].describe()
```

```python
# 計算 NAME_CONTRACT_TYPE各組 Targer 的平均
app_train.groupby(['NAME_CONTRACT_TYPE'])['TARGET'].mean()
```

```python
# 取前 10000 筆作範例: 分別將 AMT_INCOME_TOTAL, AMT_CREDIT, AMT_ANNUITY 的值除以根據 NAME_CONTRACT_TYPE 分組後的平均數。
# x : 分組的三個欄位
app_train.loc[0:10000, ['NAME_CONTRACT_TYPE', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY']].groupby(['NAME_CONTRACT_TYPE']).apply(lambda x: x / x.mean())
```

```python
app_train.loc[0:10000, ['NAME_CONTRACT_TYPE', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY']].groupby(['NAME_CONTRACT_TYPE']).apply(lambda x: print(x))
```

```python
# 分組看 tartget 的 target 的 histogram
app_train.groupby(['NAME_CONTRACT_TYPE'])['TARGET'].hist()
plt.show()
```

## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩<br><br>
    [pd.cut參考](https://medium.com/@morris_tai/pandas%E7%9A%84cut-qcut%E5%87%BD%E6%95%B8-93c244e34cfc)

```python
#1
cut_rule = [0, 1, 3, 6, np.inf]

app_train['CNT_CHILDREN_GROUP'] = pd.cut(app_train['CNT_CHILDREN'].values, cut_rule, include_lowest=True, right = False, labels=['no child', '1~2 children', '3~5 children', 'more than 5 children'])
app_train['CNT_CHILDREN_GROUP'].value_counts()
```

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot

```python
#2-1
#grouped_df = app_train.groupby(['CNT_CHILDREN_GROUP','TARGET'])['AMT_INCOME_TOTAL'].mean()
#print(grouped_df)

grp = ['CNT_CHILDREN_GROUP','TARGET']

grouped_df = app_train.groupby(grp)['AMT_INCOME_TOTAL']
grouped_df.mean()
```

```python
#2-2
plt_column = ['AMT_INCOME_TOTAL']
plt_by = ['CNT_CHILDREN_GROUP','TARGET']

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False, figsize=(12,12))
plt.suptitle('')
plt.show()
```

3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
grouped_df.apply(lambda x: x).head()
```

```python
#3
app_train['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET'] = grouped_df.apply(lambda x: ( x - np.mean(x) ) / ( np.std(x) ))

app_train[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head()
```
